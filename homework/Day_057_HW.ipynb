# 作業
- 作業目標 : 用 iris (dataset.load_iris()) 資料嘗試跑 hierarchical clustering

```python
from sklearn import datasets

iris = datasets.load_iris()
X = iris.data
y = iris.target
```

載入 相關套件 並 執行 hierarchical clustering 實驗 ...

```python
from sklearn.cluster import AgglomerativeClustering
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 6))
plt.subplot(121)
plt.scatter(X[:, 0], X[:, 1], c=y)
plt.title('sepal length:x - width:y')

plt.subplot(122)
plt.scatter(X[:, 2], X[:, 3], c=y)
plt.title('petal length:x - width:y')

# 三組,使用最遠距離
clf = AgglomerativeClustering(n_clusters=3, linkage='complete')

clustering = clf.fit(X)

y_pred = clustering.labels_

plt.figure(figsize=(12, 6))
plt.subplot(121)
plt.scatter(X[:, 0], X[:, 1], c=y_pred)
plt.title('sepal length:x - width:y')

plt.subplot(122)
plt.scatter(X[:, 2], X[:, 3], c=y_pred)
plt.title('petal length:x - width:y_pred')
```

```python
plt.figure(figsize=(12, 6))
plt.subplot(121)
plt.scatter(X[:, 0], X[:, 1], c=y)
plt.title('sepal length:x - width:y')

plt.subplot(122)
plt.scatter(X[:, 2], X[:, 3], c=y)
plt.title('petal length:x - width:y')

# 三組,使用平均距離
clf = AgglomerativeClustering(n_clusters=3, linkage='average')

clustering = clf.fit(X)

y_pred = clustering.labels_

plt.figure(figsize=(12, 6))
plt.subplot(121)
plt.scatter(X[:, 0], X[:, 1], c=y_pred)
plt.title('sepal length:x - width:y')

plt.subplot(122)
plt.scatter(X[:, 2], X[:, 3], c=y_pred)
plt.title('petal length:x - width:y_pred')
```

就以上兩個例子來看，似乎以average distance所得到分組結果較符合原始資料分布
