- 作業目標 : iris (dataset.load_iris()) 資料嘗試跑 kmeans (可以測試不同的群數 , init 等)

```python
import matplotlib.pyplot as plt
from sklearn import datasets

iris = datasets.load_iris()
X = iris.data
y = iris.target
```

### 載入 相關套件 並 執行 kmean 實驗 ...

- 實驗測試不同的群數
- 實驗測試不同的初始值
- 呈現結果

```python
from sklearn.cluster import KMeans

def try_kmeans(clusters:int, init:str):
    kmeans = KMeans(n_clusters=clusters, init=init)
    kmeans.fit(X)
    y_pred = kmeans.labels_

    plt.figure(figsize=(12, 6))
    plt.subplot(121)
    plt.scatter(X[:, 0], X[:, 1], c=y_pred)
    plt.title('sepal length:x - width:y')

    plt.subplot(122)
    plt.scatter(X[:, 2], X[:, 3], c=y_pred)
    plt.title('petal length:x - width:y')
```

```python
# 三群, k-means++
try_kmeans(3, 'k-means++')

# 三群, random
try_kmeans(3, 'random')
```

```python
# 四群, k-means++
try_kmeans(4, 'k-means++')

# 四群, random
try_kmeans(4, 'random')
```
