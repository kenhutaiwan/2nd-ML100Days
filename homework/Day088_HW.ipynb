```python
import os
import keras

# Disable GPU
os.environ["CUDA_VISIBLE_DEVICES"] = ""

# 預防錯誤： OMP: Error #15: Initializing libiomp5.dylib, but found libiomp5.dylib already initialized.
os.environ['KMP_DUPLICATE_LIB_OK'] = 'True'
```

```python
train, test = keras.datasets.cifar10.load_data()
```

```python
## 資料前處理
def preproc_x(x, flatten=True):
    x = x / 255.
    if flatten:
        x = x.reshape((len(x), -1))
    return x

def preproc_y(y, num_classes=10):
    if y.shape[-1] == 1:
        y = keras.utils.to_categorical(y, num_classes)
    return y    
```

```python
x_train, y_train = train
x_test, y_test = test

# Preproc the inputs
x_train = preproc_x(x_train)
x_test = preproc_x(x_test)

# Preprc the outputs
y_train = preproc_y(y_train)
y_test = preproc_y(y_test)
```

```python
from keras.layers import BatchNormalization

# 三層隱藏層，一個輸出層(softmax)
def build_mlp(input_shape, output_units=10, drp_ratio=0.2):
    input_layer = keras.layers.Input(input_shape)
    x = keras.layers.Dense(units=512, activation="relu", name="hidden_layer_0")(input_layer)
    x = BatchNormalization()(x)
    x = keras.layers.Dense(units=256, activation="relu", name="hidden_layer_1")(x)
    x = BatchNormalization()(x)
    x = keras.layers.Dense(units=128, activation="relu", name="hidden_layer_2")(x)
    x = BatchNormalization()(x)
    out = keras.layers.Dense(units=output_units, activation="softmax", name="output")(x)
    model = keras.models.Model(inputs=[input_layer], outputs=[out])
    
    return model
```

```python
## 超參數設定
LEARNING_RATE = 1e-3
EPOCHS = 50
BATCH_SIZE = 1024
MOMENTUM = 0.95
```

```python
# 載入 Callbacks
from keras.callbacks import Callback

# Record_fp_tp
class Record_tp_tn(Callback):
    """Code Here
    撰寫一個紀錄 ture positive 與 true negative 數量的 callbacks    
    https://zhuanlan.zhihu.com/p/51356820 這裡或許有線索
    """
    def on_epoch_end(self, epoch, logs = {}, thres=0.5):
        logs = logs or {}
        y_pred = self.model.predict(self.validation_data[0])

        for i in range(10):
            logs['y_true_{}'.format(str(i))] = (self.validation_data[1][:, i] > 0) * 1
            logs['y_pred_{}'.format(str(i))] = (y_pred[:, i] >= thres) * 1            
            logs['measure_{}'.format(str(i))] = self.perf_measure(logs['y_true_{}'.format(str(i))] , logs['y_pred_{}'.format(str(i))])
#         y_true = self.validation_data[1].argmax(axis = 1)
#         y_pred = self.model.predict(self.validation_data[0])
#         y_pred = y_pred.argmax(axis = 1)
        

    # see https://stackoverflow.com/questions/31324218/scikit-learn-how-to-obtain-true-positive-true-negative-false-positive-and-fal
    def perf_measure(self, y_actual, y_hat):
        TP = 0
        FP = 0
        TN = 0
        FN = 0

        for i in range(len(y_hat)): 
            if y_actual[i]==y_hat[i]==1:
               TP += 1
            if y_hat[i]==1 and y_actual[i]!=y_hat[i]:
               FP += 1
            if y_actual[i]==y_hat[i]==0:
               TN += 1
            if y_hat[i]==0 and y_actual[i]!=y_hat[i]:
               FN += 1
        return (TP, FP, TN, FN)       
        
rec_tptn = Record_tp_tn()
```

```python
model = build_mlp(input_shape=x_train.shape[1:])
model.summary()
optimizer = keras.optimizers.SGD(lr=LEARNING_RATE, nesterov=True, momentum=MOMENTUM)
model.compile(loss="categorical_crossentropy", metrics=["accuracy"], optimizer=optimizer)

"""Code Here
將 callbacks 放入模型訓練流程中
"""
model.fit(x_train, y_train, 
          epochs=EPOCHS, 
          batch_size=BATCH_SIZE, 
          validation_data=(x_test, y_test), 
          shuffle=True,
          callbacks=[rec_tptn]
         )

# Collect results
train_loss = model.history.history["loss"]
valid_loss = model.history.history["val_loss"]
train_acc = model.history.history["acc"]
valid_acc = model.history.history["val_acc"]

```

```python
"""Code Here
將 tp/tn 從訓練完的模型中取出
"""
# 以手寫數字9為例
measures = model.history.history['measure_9']
valid_tp = [t[0] for t in measures]
valid_tn = [t[1] for t in measures]
```

```python
import matplotlib.pyplot as plt
%matplotlib inline

plt.plot(range(len(train_loss)), train_loss, label="train loss")
plt.plot(range(len(valid_loss)), valid_loss, label="valid loss")
plt.legend()
plt.title("Loss")
plt.show()

plt.plot(range(len(train_acc)), train_acc, label="train accuracy")
plt.plot(range(len(valid_acc)), valid_acc, label="valid accuracy")
plt.legend()
plt.title("Accuracy")
plt.show()

plt.plot(range(len(valid_tp)), valid_tp, label="valid tp", color="navy")
plt.plot(range(len(valid_tn)), valid_tn, label="valid tn", color="red")
plt.legend()
plt.title("True positives and True Negatives")
plt.show()
```
