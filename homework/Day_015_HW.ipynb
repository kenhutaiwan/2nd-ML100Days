````python
# 建立一個 10 x 10, 數值分布自 -1.0 ~ 1.0 的矩陣並繪製 Heatmap

matrix = np.random.uniform(low=-1.0, high=1.0, size=(10, 10))

plt.figure(figsize=(10,10))

heatmap = sns.heatmap(matrix, vmin=-1.0, vmax=1.0, center=0.0, annot=True, linewidths=0.2)

plt.show()
````

### numpy產生常態分佈數值的矩陣
```
numpy.random.normal(loc=0.0, scale=1.0, size=None)
```

````python
# 建立一個 1000 x 3, 數值分布為 -1.0 ~ 1.0 的矩陣，並繪製 PairPlot (上半部為 scatter, 對角線為 hist, 下半部為 density)

nrow = 1000
ncol = 3
matrix = np.random.uniform(low=-1.0, high=1.0, size=(nrow, ncol)) 

# 隨機給予 0, 1, 2 三種標籤
indice = np.random.choice([0,1,2], size=nrow)
plot_data = pd.DataFrame(matrix, indice)

# 繪製 seborn 進階 Heatmap
grid = sns.PairGrid(data = plot_data, size = 3, diag_sharey=False)

grid.map_upper(plt.scatter , alpha = 0.2)
grid.map_diag(plt.hist)
grid.map_lower(sns.kdeplot , cmap = plt.cm.OrRd_r)

plt.show()
````

````python
# 建立一個 1000 x 3, 數值分布為常態分佈的矩陣，並繪製 PairPlot (上半部為 scatter, 對角線為 hist, 下半部為 density)

nrow = 1000
ncol = 3

matrix = np.random.normal(size=(nrow, ncol))

# 隨機給予 0, 1, 2 三種標籤
indice = np.random.choice([0,1,2], size=nrow)
plot_data = pd.DataFrame(matrix, indice)

# 繪製 seborn 進階 Heatmap
grid = sns.PairGrid(data = plot_data, size = 3, diag_sharey=False)

grid.map_upper(plt.scatter , alpha = 0.2)
grid.map_diag(plt.hist)
grid.map_lower(sns.kdeplot , cmap = plt.cm.OrRd_r)

plt.show()
````
